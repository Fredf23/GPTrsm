> *У нас получился результат, но выскочили предупреждения. Они появляются, когда вы пытаетесь отобразить массив данных, который имеет значения за пределами допустимого диапазона для imshow с RGB-данными ([0…1] для float или [0…255] для int). Это означает, что matplotlib обрезает ваши данные до этого диапазона, что может привести к потере информации или искажению цветов. Чтобы избежать этого предупреждения, вам нужно либо привести ваши данные к типу uint8, либо нормализовать их в диапазоне [0…1].*

```python
# Создаем массивы размером 100х100х3 для каждого цвета
# Для белого цвета все элементы равны 255
white = np.full((100, 100, 3), 255)
# Для черного цвета все элементы равны 0
black = np.zeros((100, 100, 3))
# Для красного цвета первый канал равен 255, остальные равны 0
red = np.zeros((100, 100, 3))
red[:, :, 0] = 255
# Для зеленого цвета второй канал равен 255, остальные равны 0
green = np.zeros((100, 100, 3))
green[:, :, 1] = 255
# Для синего цвета третий канал равен 255, остальные равны 0
blue = np.zeros((100, 100, 3))
blue[:, :, 2] = 255

# Приводим массивы к типу uint8, чтобы избежать предупреждения
white = white.astype(np.uint8)
black = black.astype(np.uint8)
red = red.astype(np.uint8)
green = green.astype(np.uint8)
blue = blue.astype(np.uint8)

# Создаем фигуру и массив подграфиков размером 1х5
fig, axes = plt.subplots(1, 5, figsize=(15, 3), sharey=True)

# Отображаем каждый массив на соответствующем подграфике
axes[0].imshow(white) # Отображаем белый массив как изображение
axes[0].set_title('White') # Добавляем заголовок
axes[1].imshow(black) # Отображаем черный массив как изображение
axes[1].set_title('Black') # Добавляем заголовок
axes[2].imshow(red) # Отображаем красный массив как изображение
axes[2].set_title('Red') # Добавляем заголовок
axes[3].imshow(green) # Отображаем зеленый массив как изображение
axes[3].set_title('Green') # Добавляем заголовок
axes[4].imshow(blue) # Отображаем синий массив как изображение
axes[4].set_title('Blue') # Добавляем заголовок

# Показываем фигуру
plt.show()
```

```python
# Загрузка изображения в переменную img
img = cv2.imread("bubbles.jpg")

# Преобразование изображения в оттенки серого
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Применение фильтра Гаусса для сглаживания шума
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Выполнение распознавания окружностей с помощью функции cv2.HoughCircles
circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1, 20, param1=70, param2=45, minRadius=10, maxRadius=70)

# Отрисовка окружностей на изображении
if circles is not None:
  circles = np.round(circles[0, :]).astype("int")
  for (x, y, r) in circles:
    cv2.circle(img, (x, y), r, (0, 255, 0), 4)
print(circles.shape[0], 'круга')
cv2_imshow(img)
```
